```python
from bs4 import BeautifulSoup
from urllib.request import urlopen  
# Request from url ?
```

```python
save_dir="images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
query="Dhoni"
response=requests.get(f"https://www.google.com/search?q={query}&tbm=isch&ved=2ahUKEwi_pt6i_s6AAxULnmMGHZSRCOoQ2-cCegQIABAA&oq=dhoni&gs_lcp=CgNpbWcQAzIICAAQgAQQsQMyCAgAEIAEELEDMgsIABCABBCxAxCDATIICAAQgAQQsQMyCAgAEIAEELEDMgUIABCABDILCAAQagAQQsQMQgwEyCAgAEIAEELEDMgUIABCABDIFCAAQgAQ6BwgAEBgQgARQ9whY9whgghFoAHAAeACAAbQBiAHgApIBAzAuMpgBAKABAaoBC2d3cy13aXotaW1nwAEB&sclient=img&ei=JzfTZP_2Iou8juMPlKOi0A4&bih=594&biw=530&rlz=1C1ONGR_enIN1058IN1058")
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
img_tag=soup.find_all("img")
```

```python
del img_tag[0]
```

```python
for i in img_tag:
    image_url=i['src']
    image_data=requests.get(image_url).content
    with open(os.path.join(save_dir,f"{query}_{img_tag.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
```
